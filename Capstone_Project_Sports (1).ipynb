This exploration of the athletes' ages points to a potential problem, however: The Art Competitions. During the earlier stages of the OlympiC Games, the competitions were not limited to athletics, instead featuring competitions in arts, i.e. architecture, literature, music, sculpturing and painting. These competitions, while certainly a notable part of the history of the games, were abandoned after 1954. They may also skew our analysis, since they contain entries like John Quincy Adams Ward, who's sculturing work was considered in the Olympic Games of 1928, 18 years after Ward's death. Notice that data like this could greatly skew the mean age, as an example. For simplicities sake I have therefore decided to simply exclude the Art Competitions from any further analysis. This should even the playing field between the games that took place before 1954 and those that took place after the decision to discontinue the Art Competitions.

Out of all 134732 athletes, only roughly one quarter was female. After reading a little about the history of the modern Olympic Games, I stronlgy suspect that this is due to female athletes being initially excluded (and only admitted slowly after that). Let us visualize this development. For that, I have to slightly adapt the dataframe by grouping by year and adding new columns containing the amount of male and female athletes competing during the games.

```python
df['Male'] = np.where(df['Sex']== 'M', 1, 0)
```

```python
df['Female'] = np.where(df['Sex']== 'F', 1, 0)
```

```python
df_sex = df[['Year', 'Male', 'Female']]
df_sex.head()
```

```python
df_sex['male_count'] = df_sex.groupby('Year')['Male'].transform('sum')
```

```python
df_sex['female_count'] = df_sex.groupby('Year')['Female'].transform('sum')
```

```python
df_sex = df_sex[['Year', 'male_count', 'female_count']]
df_sex_grouped = df_sex.groupby('Year').mean()
sex_plot_df = pd.DataFrame(df_sex_grouped)
sex_plot_df.head()
```

This should do. Now I can plot the distribution of athletes gender over time.

```python
sns.set(style='white')

sns.set(rc = {'figure.figsize':(20,12)})
g = sex_plot_df.plot(kind='bar', stacked=True, color=['navy', 'deepskyblue'])
g.set(xlabel = 'Year')
g.legend(loc='upper right', labels=['Male Athletes', 'Female Athletes'])
g.grid(b=None)
#g.figure.savefig('gender_time.png',edgecolor='none', transparent=True)
#files.download('gender_time.png')
```

As assumed we can see that the nubmer of female competitors was significantly lower than that of their male counterparts for quite some time. In fact,there were hardly any female competitors during the earlier years of the games. Only recently did the number of female athletes come closer to that of the males, which serves as a satisfactory explanation for the skew in the distribution of genders across the total number of competitors we observed above. 

Another bit of history to discover in this plot is the split of winter and summer games that took place after 1992. Until then both winter and summer games had been held together every four years. Starting in 1994, winter and summer games were held separately. As compensation for this change, we can at least claim that there is some olympic event every two years now. 

Another interesting topic to explore is the number of sports and events and how both of them changed over time. As we can see in the plot above, the number of athletes increased constantly, until the games were split into summer and winter games in 1994, so a reasonable assumption would be that both the number events and sports rose in turn.

Now for the final and probably most interesting hypothesis: The connection between competing at home and winning a medal. Are the chances higher to actually place on the podium when an athlete is competing in his or her homecountry?

For this, I will have to add a 'Location_games' column, which I can then match with the NOC of the athlete. This has to be done manually, for which I will use a function that assignes values for Country depending on which city hosted the games. 

```python
df.Team.unique()
```

```python
def f(row):
    if row['City'] == 'Barcelona':
        val = 'ESP'
    elif row['City'] == 'London':
        val = 'GBR'
    elif row['City'] == 'Antwerpen':
        val = 'NED'
    elif row['City'] == 'Paris':
        val = 'FRA'
    elif row['City'] == 'Calgary':
        val = 'CAN'
    elif row['City'] == 'Albertville':
        val = 'FRA'
    elif row['City'] == 'Lillehammer':
        val = 'NOR'
    elif row['City'] == 'Salt Lake City':
        val = 'USA'
    elif row['City'] == 'Los Angeles':
        val = 'USA'
    elif row['City'] == 'Helsinki':
        val = 'FIN'
    elif row['City'] == 'Lake Placid':
        val = 'USA'
    elif row['City'] == 'Sydney':
        val = 'AUS'
    elif row['City'] == 'Atlanta':
        val = 'USA'
    elif row['City'] == 'Stockholm':
        val = 'SWE'
    elif row['City'] == 'Sochi':
        val = 'RUS'
    elif row['City'] == 'Nagano':
        val = 'JPN'
    elif row['City'] == 'Torino':
        val = 'ITA'
    elif row['City'] == 'Beijing':
        val = 'CHN'
    elif row['City'] == 'Rio de Janeiro':
        val = 'BRA'
    elif row['City'] == 'Athina':
        val = 'GRE'
    elif row['City'] == 'Squaw Valley':
        val = 'USA'
    elif row['City'] == 'Innsbruck':
        val = 'AUT'
    elif row['City'] == 'Sarajevo':
        val = 'BIH'
    elif row['City'] == 'Mexico City':
        val = 'MEX'
    elif row['City'] == 'Munich':
        val = 'GER'
    elif row['City'] == 'Seoul':
        val = 'KOR'
    elif row['City'] == 'Berlin':
        val = 'GER'
    elif row['City'] == 'Oslo':
        val = 'NOR'
    elif row['City'] == 'Cortina d\'Ampezzo':
        val = 'ITA'
    elif row['City'] == 'Melbourne':
        val = 'AUS'
    elif row['City'] == 'Roma':
        val = 'ITA'
    elif row['City'] == 'Montreal':
        val = 'CAN'
    elif row['City'] == 'Amsterdam':
        val = 'NED'
    elif row['City'] == 'Moskva':
        val = 'RUS'
    elif row['City'] == 'Tokyo':
        val = 'JPN'
    elif row['City'] == 'Vancouver':
        val = 'CAN'
    elif row['City'] == 'Grenoble':
        val = 'FRA'
    elif row['City'] == 'Sapporo':
        val = 'JPN'
    elif row['City'] == 'Chamonix':
        val = 'FRA'
    elif row['City'] == 'St. Louis':
        val = 'USA'
    elif row['City'] == 'Sankt Moritz':
        val = 'SUI'
    elif row['City'] == 'Garmisch-Partenkirchen':
        val = 'GER'
    else:
        val = 'SCHEISSE'
    return val
```

```python
df['Location_games'] = df.apply(f, axis=1)
```
